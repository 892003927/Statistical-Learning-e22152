1. Proof of Finite Sample Bias in $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$

We begin by defining the sample deviances from the true population mean $\boldsymbol{\mu}$:$$\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n = (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$$Substituting this expression into the Maximum Likelihood Estimation formula yields:$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right]^T$$Expanding the matrix multiplication:$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - (\mathbf{X}_i - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T + (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T \right]$$Distributing the sum over the individual terms, and noting that $\sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu}) = n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$:$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - \frac{n}{n} (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$Taking the expectation $\mathbb{E}[\cdot]$ on both sides:$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} \sum_{i=1}^n \mathbb{E}\left[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T\right] - \mathbb{E}\left[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T\right]$$Since $\mathbf{X}_i$ are i.i.d., $\mathbb{E}\left[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T\right] = \boldsymbol{\Sigma}$.The variance of the sample mean is $\text{Var}(\widehat{\boldsymbol{\mu}}_n) = \mathbb{E}\left[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T\right] = \frac{1}{n}\boldsymbol{\Sigma}$.Plugging these parameters back into the equation:$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} (n\boldsymbol{\Sigma}) - \frac{1}{n}\boldsymbol{\Sigma} = \boldsymbol{\Sigma} - \frac{1}{n}\boldsymbol{\Sigma} = \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma}$$

Part C: Numerical Verification (MANOVA & Bartlett's Framework)

Instead of external references, the complete, production-grade MANOVA splitting engine using numerical tracking formulas is implemented below.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Structural drift injection in the final chunk (indices 4000-5000)
base_data[4000:, 0] += 0.015
base_data[4000:, 2] -= 0.010

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and computes Wilks' Lambda
    along side Bartlett's Chi-Square asymptotic approximation.
    """
    X = df.to_numpy()
    n, m = X.shape
    chunk_size = n // g_chunks

    # 1. Global Metrics
    global_mean = np.mean(X, axis=0)

    # Initialize Core Matrices
    W = np.zeros((m, m))
    B = np.zeros((m, m))

    # 2. Iterate through chunks to accumulate variations
    for g in range(g_chunks):
        chunk = X[g*chunk_size : (g+1)*chunk_size]
        chunk_mean = np.mean(chunk, axis=0)

        # Within-Chunk Sum of Squares
        deviations_w = chunk - chunk_mean
        W += deviations_w.T @ deviations_w

        # Between-Chunk Sum of Squares
        deviation_b = (chunk_mean - global_mean).reshape(-1, 1)
        B += chunk_size * (deviation_b @ deviation_b.T)

    # 3. Calculate Wilks' Lambda
    wilks_lambda = np.linalg.det(W) / np.linalg.det(B + W)

    # 4. Bartlett Transformation
    df_bartlett = m * (g_chunks - 1)
    chi2_calc = - (n - 1 - (m + g_chunks) / 2.0) * np.log(wilks_lambda)
    p_value = 1.0 - stats.chi2.cdf(chi2_calc, df_bartlett)

    conclusion = "Reject H0: First moment is heterogeneous (Baseline Center Shifted)" if p_value < 0.05 else "Fail to Reject H0: First moment is homogeneous"

    return {
        "Wilks' Lambda": wilks_lambda,
        "Bartlett Chi2": chi2_calc,
        "p-value": p_value,
        "Degrees of Freedom": df_bartlett,
        "Diagnostic Conclusion": conclusion
    }

# Execute alternative pipeline implementation
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
for key, val in results.items():
    print(f"{key}: {val}")
```

3. Production Engine: Factor Analysis & Visualization Grid

The code below implements the complete Factor Analysis parsing pipeline, including Varimax coordinate rotations, Thomson score generation, and a fully configured visualization grid layout using matplotlib/seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import FactorAnalysis

# --- Pipeline Instantiation ---
def run_factor_analysis_pipeline(df: pd.DataFrame, k_factors: int = 2):
    n, m = df.shape

    # 1. Z-Score Mapping Layer with Guardrails
    means = df.mean(axis=0)
    stds = df.std(axis=0, ddof=1)
    stds[stds < 1e-15] = 1e-15  # Cap zero-variance channels

    Z = (df - means) / stds

    # 2. Dimensionality Constraints
    if n <= m:
        raise ValueError("Snapshot count n must be greater than feature count m.")
    if k_factors >= m:
        raise ValueError("Latent factors k must be strictly less than feature dimension m.")

    # 3. Fit Factor Analysis Model (includes implicit Varimax rotation fallback/activation)
    fa = FactorAnalysis(n_components=k_factors, rotation='varimax', random_state=42)
    fa.fit(Z)

    # Extract structural configuration matrices
    loadings = fa.components_.T  # Shape: (m, k)
    uniqueness = fa.noise_variance_
    communality = 1.0 - uniqueness

    # 4. Latent Score Projection via Thomson Regression Method
    R_emp = np.corrcoef(Z.T)
    R_inv = np.linalg.inv(R_emp)
    # Thomson Score Mapping weights
    F = Z.to_numpy() @ R_inv @ loadings

    # --- Generate Production-Grade 2x2 Dashboard Canvas ---
    fig, axes = plt.subplots(2, 2, figsize=(13, 8))
    plt.subplots_adjust(hspace=0.35, wspace=0.3)

    # Subplot (1,1): Structural Loadings Matrix Heatmap
    sns.heatmap(np.abs(loadings), annot=True, cmap="YlOrRd", cbar=True,
                yticklabels=df.columns, xticklabels=[f"Factor {i+1}" for i in range(k_factors)],
                ax=axes[0,0])
    axes[0,0].set_title("Absolute Rotated Loadings Matrix")

    # Subplot (1,2): Variance Partitioning Profile
    sensor_indices = np.arange(m)
    axes[0,1].barh(sensor_indices, communality * 100, label="Shared Communality (h²)", color="#1f77b4")
    axes[0,1].barh(sensor_indices, uniqueness * 100, left=communality * 100, label="Isolated Uniqueness (φ²)", color="#ff7f0e")
    axes[0,1].set_yticks(sensor_indices)
    axes[0,1].set_yticklabels(df.columns)
    axes[0,1].set_xlim(0, 100)
    axes[0,1].set_title("Variance Partitioning Profile (%)")
    axes[0,1].legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)

    # Subplot (2,1): Sensor Uniqueness Line Profile
    axes[1,0].plot(df.columns, uniqueness, r'--x', color="#d62728", linewidth=2, markersize=8)
    axes[1,0].set_title("Sensor Noise Floor Profile (Uniqueness)")
    axes[1,0].set_xticklabels(df.columns, rotation=25, ha="right")
    axes[1,0].grid(True, linestyle=":", alpha=0.6)

    # Subplot (2,2): Reconstructed Empirical Subspace Variance
    factor_vars = np.var(F, axis=0, ddof=1)
    axes[1,2 if False else 1].bar([f"Factor {i+1}" for i in range(k_factors)], factor_vars, color="#2ca02c", edgecolor='black', linewidth=0.5)
    axes[1,1].set_title("Latent Factor Empirical Variances")
    axes[1,1].grid(True, axis='y', linestyle=":", alpha=0.6)

    plt.suptitle("Universal Latent Space Subspace Diagnostics Canvas", fontsize=14, y=0.98)
    return fig, F

# Execute optimization loop over dummy sensor asset architecture
np.random.seed(101)
dummy_data = np.random.normal(loc=0, scale=1, size=(3000, 4))
df_sensors = pd.DataFrame(dummy_data, columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4'])

fig, factor_scores = run_factor_analysis_pipeline(df_sensors, k_factors=2)
plt.show()
```
